```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')


df_confirmed_2019 = pd.read_csv("data/COVID 19-3-2020/time_series_2019-ncov-Confirmed.csv")
df_deaths_2019 = pd.read_csv("data/COVID 19-3-2020/time_series_2019-ncov-Deaths.csv")
df_recovered_2019 = pd.read_csv("data/COVID 19-3-2020/time_series_2019-ncov-Recovered.csv")
df_confirmed = pd.read_csv("data/COVID 19-3-2020/time_series-ncov-Confirmed.csv")
df_deaths = pd.read_csv("data/COVID 19-3-2020/time_series-ncov-Deaths.csv")
df_recovered = pd.read_csv("data/COVID 19-3-2020/time_series-ncov-Recovered.csv")
```

```python
class Timeseries:
    def __init__(self, df):
        self.df = df
    def plot(self, country):
#         plt.title(country)
        plt.plot(self.df.loc[self.df['Country/Region'] == country].iloc[0][4:], '.', label=country)
        plt.legend(loc='upper left')
        plt.show()
```

```python
ts_confirmed_2019 = Timeseries(df_confirmed_2019)
ts_confirmed_2019.plot("Singapore")
ts_confirmed_2019.plot("Malaysia")
ts_confirmed_2019.plot("Indonesia")
```

```python
ts_recovered_2019 = Timeseries(df_recovered_2019)
ts_recovered_2019.plot("Singapore")
ts_recovered_2019.plot("Malaysia")
ts_recovered_2019.plot("Indonesia")
```
